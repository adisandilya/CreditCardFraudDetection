```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from  sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,precision_score,recall_score,f1_score,classification_report
```

```python
#loading the data set
cc=pd.read_csv('/content/creditcard.csv')
```

```python
#visualizing the data
LABELS = ["Normal Transation ","Fraud Transaction"]
count_classes = pd.value_counts(cc['Class'], sort = True)
count_classes.plot(kind = 'bar', rot=0)
plt.xticks(range(2), LABELS)
plt.title("Transaction Class Distribution")
plt.xlabel("Class")
plt.ylabel("No of Occurance")
plt.show()
```

```python
# separating the data for analysis
Normal = cc[cc.Class == 0]
fraud = cc[cc.Class == 1]
print(Normal.shape)
print(fraud.shape)
```

```python
Normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
cc.groupby('Class').mean()
```

```python
#Under-Sampling

#Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions
```

```python
ls=Normal.sample(n=492)
```

```python
#Concatinating two data frames
nd=pd.concat([ls,fraud],axis=0)
```

```python
nd.head()
```

```python
nd.tail()
```

```python
nd['Class'].value_counts()
```

```python
nd.groupby('Class').mean()
```

```python
#splitting the data
X=nd.drop(columns='Class',axis=1)
Y=nd['Class']
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training & testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model training
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
#model Evaluation
result = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:")
print(result)
print("Accuracy:",accuracy_score(Y_test, Y_pred))
print("Report: ")
print(classification_report(Y_test, Y_pred))
print("Precision:",precision_score(Y_test, Y_pred))
print("Recall:",recall_score(Y_test, Y_pred))
print("F1 Score:",f1_score(Y_test, Y_pred))
```
